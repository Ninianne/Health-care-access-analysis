```python
import pandas as pd

# Correct relative path based on ZIP structure
df = pd.read_csv('SVI_2022_US.csv.download/SVI_2022_US.csv')

# Preview the data
df.head()
```

```python
# Select key columns relevant to healthcare access
df_health = df[[
    'FIPS', 'STATE', 'ST_ABBR', 'COUNTY',
    'E_POV', 'E_UNEMP', 'E_NOHSDP', 'E_AGE65',
    'E_DISABL', 'E_NOVEH', 'E_GROUPQ',
    'RPL_THEME1', 'RPL_THEME2', 'RPL_THEME4', 'RPL_THEMES'
]]

# Rename for clarity
df_health.rename(columns={
    'E_POV': 'Pct_Poverty',
    'E_UNEMP': 'Pct_Unemployed',
    'E_AGE65': 'Pct_Age_65plus',
    'E_DISABL': 'Pct_Disabled',
    'E_NOVEH': 'Pct_NoVehicle',
    'E_NOHSDP': 'Pct_No_HS_Diploma',
    'E_GROUPQ': 'Pct_GroupQuarters',
    'RPL_THEME1': 'Score_Socioeconomic',
    'RPL_THEME2': 'Score_Household',
    'RPL_THEME4': 'Score_Housing_Transport',
    'RPL_THEMES': 'SVI_Overall'
}, inplace=True)

# Preview cleaned data
df_health.head()
```

```python
# Select relevant healthcare-related columns
df_health = df[[
    'FIPS', 'STATE', 'ST_ABBR', 'COUNTY',
    'E_POV150',        # % Below 150% of poverty
    'E_UNEMP',         # % Unemployed
    'E_NOHSDP',        # % No High School Diploma
    'E_AGE65',         # % Age 65+
    'E_DISABL',        # % Disabled
    'E_NOVEH',         # % No Vehicle Access
    'E_GROUPQ',        # % Group Quarters
    'RPL_THEME1',      # Socioeconomic Status Score
    'RPL_THEME2',      # Household Composition
    'RPL_THEME4',      # Housing & Transportation
    'RPL_THEMES'       # Overall Vulnerability Index
]]

# Rename for clarity
df_health.rename(columns={
    'E_POV150': 'Pct_Poverty',
    'E_UNEMP': 'Pct_Unemployed',
    'E_NOHSDP': 'Pct_No_HS_Diploma',
    'E_AGE65': 'Pct_Age_65plus',
    'E_DISABL': 'Pct_Disabled',
    'E_NOVEH': 'Pct_NoVehicle',
    'E_GROUPQ': 'Pct_GroupQuarters',
    'RPL_THEME1': 'Score_Socioeconomic',
    'RPL_THEME2': 'Score_Household',
    'RPL_THEME4': 'Score_Housing_Transport',
    'RPL_THEMES': 'SVI_Overall'
}, inplace=True)

# Preview cleaned data
df_health.head()
```

```python
# Summary statistics
df_health.describe()
```

```python
# Sort by overall social vulnerability score (highest to lowest)
top_vulnerable = df_health.sort_values(by='SVI_Overall', ascending=False).head(10)
top_vulnerable[['STATE', 'COUNTY', 'SVI_Overall']]
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
plt.scatter(df_health['Pct_Poverty'], df_health['Pct_NoVehicle'], alpha=0.5)
plt.title('Poverty vs. Lack of Vehicle Access')
plt.xlabel('% Below Poverty')
plt.ylabel('% No Vehicle Access')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(df_health['Pct_Age_65plus'], df_health['Score_Housing_Transport'], alpha=0.5, color='orange')
plt.title('% Age 65+ vs. Housing & Transportation Score')
plt.xlabel('% Age 65+')
plt.ylabel('Housing & Transportation Vulnerability Score')
plt.grid(True)
plt.show()
```

```python
df_elderly_transport = df_health.sort_values(
    by=['Pct_Age_65plus', 'Pct_NoVehicle'], ascending=False
).head(10)

df_elderly_transport[['STATE', 'COUNTY', 'Pct_Age_65plus', 'Pct_NoVehicle']]
```

```python
# Normalize values (0 to 1) so we can combine meaningfully
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df_health[['Norm_Poverty', 'Norm_Age65', 'Norm_NoVehicle']] = scaler.fit_transform(
    df_health[['Pct_Poverty', 'Pct_Age_65plus', 'Pct_NoVehicle']]
)

# Create a composite vulnerability score (adjust weights if needed)
df_health['Health_Access_Barrier_Score'] = (
    0.4 * df_health['Norm_Poverty'] +
    0.3 * df_health['Norm_Age65'] +
    0.3 * df_health['Norm_NoVehicle']
)

# Preview
df_health[['STATE', 'COUNTY', 'Health_Access_Barrier_Score']].sort_values(
    by='Health_Access_Barrier_Score', ascending=False
).head(10)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
sns.scatterplot(
    x=df_health['Health_Access_Barrier_Score'],
    y=df_health['SVI_Overall']
)
plt.title('Custom Barrier Score vs. CDC Social Vulnerability Index')
plt.xlabel('Health Access Barrier Score')
plt.ylabel('SVI Overall')
plt.grid(True)
plt.show()
```

```python
top_barrier_counties = df_health.sort_values(
    by='Health_Access_Barrier_Score', ascending=False
).head(10)

top_barrier_counties[['STATE', 'COUNTY', 'Pct_Poverty', 'Pct_Age_65plus', 'Pct_NoVehicle', 'Health_Access_Barrier_Score']]
```
